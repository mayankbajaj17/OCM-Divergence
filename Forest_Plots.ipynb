```python
import matplotlib.pyplot as plt
import os
from google.colab import drive

# --- 1. Mount Google Drive ---
# This will prompt you to authorize access to your Drive.
drive.mount('/content/drive')

# --- 2. Define Output Path ---
# We define the specific folder path you requested.
output_folder = '/content/drive/MyDrive/Figure 4 Manuscript/'

# Check if directory exists; if not, notify the user (or create it)
if not os.path.exists(output_folder):
    print(f"Note: The folder '{output_folder}' was not found. Creating it now...")
    os.makedirs(output_folder)
else:
    print(f"Success: Found folder '{output_folder}'.")

# --- 3. Data Setup (Sourced from provided tables) ---
# Format: [Label, HR, Lower_CI, Upper_CI, P_value_string]

# TCGA Data
data_tcga = [
    ["IDH encoded", 7.319, 5.197, 10.31, "< 0.001"],
    ["Combined Score\n(Z-Filtered)", 4.498, 1.94, 10.43, "0.0005"]
]

# CGGA Data
data_cgga = [
    ["IDH encoded", 3.052, 2.576, 3.618, "< 0.001"],
    ["Combined Score\n(Z-Filtered)", 1.396, 1.142, 1.707, "0.0011"]
]

# --- 4. Plotting Function ---
def draw_forest_panel(ax, data, title, color_marker):
    # Layout Limits
    ax.set_ylim(-0.8, len(data) - 0.2)
    ax.set_xlim(0, 14)

    # Hide Spines
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    # Reference Line
    ax.axvline(x=1, color='grey', linestyle='--', linewidth=1)

    for i, row in enumerate(data):
        label, hr, lower, upper, p_val = row
        y = i

        # Plot Error Bar & Point
        ax.plot([lower, upper], [y, y], color='black', linewidth=1.5, solid_capstyle='round')
        ax.plot(hr, y, marker='s', color=color_marker, markersize=8, markeredgecolor='black')

        # Text Columns (Manually positioned for alignment)
        # Covariate Label
        ax.text(-6.0, y, label, ha='left', va='center', fontsize=12, fontfamily='serif', weight='bold')
        # HR Value
        ax.text(15.0, y, f"{hr:.2f} ({lower:.2f}-{upper:.2f})", ha='left', va='center', fontsize=12, fontfamily='serif')
        # P-value
        ax.text(22.0, y, p_val, ha='center', va='center', fontsize=12, fontfamily='serif')

    # Headers (placed above the first row)
    header_y = len(data) - 0.2
    ax.text(-6.0, header_y, "Covariate", ha='left', va='bottom', fontsize=12, fontweight='bold', fontfamily='serif')
    ax.text(15.0, header_y, "Hazard Ratio (95% CI)", ha='left', va='bottom', fontsize=12, fontweight='bold', fontfamily='serif')
    ax.text(22.0, header_y, "P-value", ha='center', va='bottom', fontsize=12, fontweight='bold', fontfamily='serif')

    # Titles & Labels
    ax.set_title(title, loc='left', fontsize=14, fontweight='bold', fontfamily='serif', x=-0.35, y=1.1)
    ax.set_xlabel("Hazard Ratio", fontsize=12, fontfamily='serif')
    ax.set_yticks([])
    ax.set_xticks([0, 2, 4, 6, 8, 10, 12, 14])

# --- 5. Generate and Save ---
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), dpi=300)
plt.subplots_adjust(left=0.3, right=0.8, hspace=0.6) # Margins for text

# Draw TCGA (Top) and CGGA (Bottom)
draw_forest_panel(ax1, data_tcga, "Figure 4D: TCGA Bivariate Cox Analysis", "#1f77b4")
draw_forest_panel(ax2, data_cgga, "Figure 4E: CGGA Bivariate Cox Analysis", "#d62728")

# Define full filename
filename = "Figure_4D_4E_Bivariate_Forest_Plot.png"
save_path = os.path.join(output_folder, filename)

# Save
plt.savefig(save_path, dpi=1200, bbox_inches='tight', pad_inches=0.2)
print(f"Processing Complete. File saved to: {save_path}")

plt.show()
```
